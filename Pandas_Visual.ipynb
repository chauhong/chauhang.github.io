# **Pandas Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Ice Cream Ratings.csv')
df= df.set_index('Date')
##Set index so when we graph it will appear in y colunm. If we dont set index, it will appear as '0,1,2,3'
```

```python
df
```

```python
print(plt.style.available)
## This one is used to change theme, style of the graph and to browse theme
```

```python
plt.style.use('fivethirtyeight')
##This is used to pick theme
```

```python
df.plot()
```

```python
df.plot(kind='line', subplots = True,title = 'Ice Cream Rating', xlabel= 'Daily Ratings', ylabel= 'Scores')
## kind is used to specific which graph we want such as line, pie. Subplots = True means that we want to separate
##each line into a separate picture so it is easier to see.
```

```python
df.plot(kind ='bar')
```

```python
df.plot(kind='bar',stacked = 'True')
```

```python
df['Flavor Rating'].plot(kind='bar')
## This only apply if we want one variable such as Flavor Rating
```

```python
df.plot.barh()
## Want bar chart but horizontal instead. Then we have to get rid of (kind ='bar)
```

```python
df.plot.barh(stacked ='True')
```

```python
df.plot.scatter()
## With scatter plot chart, we have to specified x and y in order for it to work. If not, then it will pop error.
```

```python
df.plot.scatter( x= 'Texture Rating', y= 'Overall Rating', s= 100,c ='Green')
```

```python
df.plot.hist()
```

```python
df.boxplot()
```

```python
df.plot.area(figsize =(10,5))
```

```python
df.plot.pie(y = 'Flavor Rating',figsize=(10,10))
```
